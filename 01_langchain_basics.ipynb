## 导入必要的模块

````python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv

# 加载环境变量
load_dotenv()

print("✅ 模块导入成功！")
````

## 核心概念1：Models（模型）

Models是LangChain与各种LLM交互的统一接口。无论使用什么模型，调用方式都是一致的。

````python
# 示例1：使用OpenAI模型
from langchain_openai import ChatOpenAI
import os

# model = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)
# 其他供应商
model = ChatOpenAI(
    base_url=os.getenv("OPENAI_API_BASE"),
    api_key=os.getenv("OPENAI_API_KEY"),
    model="deepseek-ai/DeepSeek-V3", temperature=0.7
)

# 直接调用模型
response = model.invoke("你好，请做一个简单的自我介绍")
print(response.content)
````

---

## 核心概念2：Prompts（提示词模板）

Prompts允许你将提示词模板化，支持变量注入，避免硬编码。

````python
# 创建提示词模板
prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个专业的{domain}助手"),
    ("user", "{question}")
])

# 使用模板生成具体的提示词
formatted = prompt.format(domain="Python编程", question="如何读取CSV文件？")
print("生成的提示词：")
print(formatted)
````

````python
# 测试不同的变量组合
test_cases = [
    {"domain": "数据分析", "question": "如何处理缺失值？"},
    {"domain": "Web开发", "question": "什么是RESTful API？"},
    {"domain": "机器学习", "question": "解释一下过拟合现象"}
]

for case in test_cases:
    print(f"\n【{case['domain']}】")
    print(prompt.format(**case))
    print("-" * 50)
````

## 核心概念3：LCEL（LangChain Expression Language）

LCEL使用管道符`|`优雅地将组件串联起来，代码像流水线一样清晰。

````python
# 创建一个简单的Chain
prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个友好的AI助手，用简洁明了的方式回答问题。"),
    ("user", "{question}")
])

# model = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)
# 其他供应商
model = ChatOpenAI(
    base_url=os.getenv("OPENAI_API_BASE"),
    api_key=os.getenv("OPENAI_API_KEY"),
    model="deepseek-ai/DeepSeek-V3", temperature=0.7
)
output_parser = StrOutputParser()

# 使用LCEL组装Chain
chain = prompt | model | output_parser

# 执行调用
response = chain.invoke({"question": "请用一句话解释什么是机器学习"})
print(response)
````

## 完整实战：构建你的第一个LangChain应用

让我们将所有概念整合起来，创建一个完整的应用。

### 步骤1：定义提示词模板

````python
prompt = ChatPromptTemplate.from_messages([
    ("system", "你是一个友好的AI助手，用简洁明了的方式回答问题。"),
    ("user", "{question}")
])

print("✅ 提示词模板创建成功")
````

### 步骤2：初始化模型

````python
# model = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.7)
# 其他供应商
model = ChatOpenAI(
    base_url=os.getenv("OPENAI_API_BASE"),
    api_key=os.getenv("OPENAI_API_KEY"),
    model="deepseek-ai/DeepSeek-V3", temperature=0.7
)
print("✅ 模型初始化成功")
````

### 步骤3：创建输出解析器

````python
output_parser = StrOutputParser()

print("✅ 输出解析器创建成功")
````

### 步骤4：使用LCEL组装Chain

````python
chain = prompt | model | output_parser

print("✅ Chain组装完成")
print("流程：输入 → prompt处理 → model生成 → output_parser解析 → 输出")
````

### 步骤5：测试应用

````python
# 测试问题列表
questions = [
    "请用一句话解释什么是机器学习",
    "Python和Java的主要区别是什么？",
    "什么是API？"
]

for i, question in enumerate(questions, 1):
    print(f"\n{'='*60}")
    print(f"问题 {i}：{question}")
    print(f"{'='*60}")
    
    response = chain.invoke({"question": question})
    print(f"回答：{response}")
````
